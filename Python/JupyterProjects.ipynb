# Coded Correspondence (String Operations)

```python
coded_message = "xuo jxuhu! jxyi yi qd unqcfbu ev q squiqh syfxuh. muhu oek qrbu je tusetu yj? y xefu ie! iudt cu q cuiiqwu rqsa myjx jxu iqcu evviuj"
```

```python
lower_alphabet = "abcdefghijklmnopqrstuvwxyz"
upper_alphabet = lower_alphabet.upper()
alphabet_length = 26
def shift_by(char, shift):
    if char.lower() not in lower_alphabet:
        return char
    index = lower_alphabet.find(char) if char in lower_alphabet else upper_alphabet.find(char)
    new_index = index - shift
    if new_index < 0:
        new_index += 26
    elif new_index > 25:
        new_index -= 26
    return lower_alphabet[new_index] if char in lower_alphabet else upper_alphabet[new_index]
```

```python
def decode(msg, shift):
    decoded_message = [shift_by(c, -shift) for c in msg]
    return "".join(decoded_message)
def encode(msg, shift):
    decoded_message = [shift_by(c, shift) for c in msg]
    return "".join(decoded_message)
```

```python
vignere_message = "txm srom vkda gl lzlgzr qpdb? fepb ejac! ubr imn tapludwy mhfbz cza ruxzal wg zztylktoikqq!"
keyword = "friends"
```

```python
def decode_vignere(msg, key):
    key_length = len(key)
    num_key = [lower_alphabet.find(k.lower()) for k in key]
    key_index = 0
    full_key = []
    for i in range(len(msg)):
        if msg[i].lower() not in lower_alphabet:
            full_key.append(0)
            continue
        full_key.append(num_key[key_index])
        key_index += 1
        if key_index == key_length:
            key_index = 0
    decoded_msg = []
    for i in range(len(msg)):
        decoded_msg.append(decode(msg[i], full_key[i]))
    return "".join(decoded_msg)
```

```python
print(decode_vignere(vignere_message, keyword))
```

```python
def encode_vignere(msg, key):
    key_length = len(key)
    num_key = [lower_alphabet.find(k.lower()) for k in key]
    key_index = 0
    full_key = []
    for i in range(len(msg)):
        if msg[i].lower() not in lower_alphabet:
            full_key.append(0)
            continue
        full_key.append(num_key[key_index])
        key_index += 1
        if key_index == key_length:
            key_index = 0
    encoded_msg = []
    for i in range(len(msg)):
        encoded_msg.append(encode(msg[i], full_key[i]))
    return "".join(encoded_msg)
```

```python
encoded_vignere = encode_vignere("This is another message. Wow! Let's see if it works...", "attempt")
print(encoded_vignere)
print(decode_vignere(encoded_vignere, "attempt"))
```

```python
conversion = {"M": 1000000,
              "B": 1000000000}

def damage_to_float(damage):
  if damage == 'Damages not recorded': return damage
  else:
    numeral = float(damage[:-1])
    magnitude = conversion[damage[-1]]
    return numeral * magnitude

damages[1], damage_to_float(damages[1])
```

```python
def create_table(names, months, years, max_sustained_winds, areas_affected, damages, deaths):
  table = {}
  for i in range(len(names)):
    subtable = {'Name': names[i], 'Month': months[i], 'Year': years[i], 'Max Sustained Wind': max_sustained_winds[i], 'Areas Affected': areas_affected[i], 'Damage': damages[i], 'Deaths': deaths[i]}
    table[names[i]] = subtable
  return table

hurricane_table = create_table(names, months, years, max_sustained_winds, areas_affected, damages, deaths)
hurricane_table["Cuba II"]
```

```python
def restructure_table_by_year(table):
  year_table = {}
  for subtable in table.values():
    subtable_year = subtable['Year']
    if subtable_year in year_table:
      year_table[subtable_year].append(subtable)
    else:
      year_table[subtable_year] = [subtable]
  return year_table

year_table = restructure_table_by_year(hurricane_table)
year_table[1932]
```

```python
def create_area_table(table):
  area_table = {}
  for subtable in table.values():
    areas = subtable['Areas Affected']
    for area in areas:
      if area in area_table:
        area_table[area] += 1
      else:
        area_table[area] = 1
  return area_table

area_table = create_area_table(hurricane_table)
"Florida", area_table["Florida"]
```

```python
def most_affected(area_table):
  current_max_value = 0
  current_max_key = ''
  for area in area_table:
    if area_table[area] > current_max_value:
      current_max_value = area_table[area]
      current_max_key = area
    elif area_table[area] == current_max_value:
      current_max_key += ", " + area
  return current_max_key, current_max_value

most_affected(area_table)
```

```python
def deadliest(hurricane_table):
  current_max_value = 0
  current_max_key = ''
  for subtable in hurricane_table.values():
    current_key = subtable["Name"]
    current_value = subtable["Deaths"]
    if current_value > current_max_value:
      current_max_value = current_value
      current_max_key = current_key
    elif current_value == current_max_value:
      current_max_key += ", " + current_key
  return current_max_key, current_max_value

deadliest(hurricane_table)
```

```python
mortality_scale = {0: 0,
                   1: 100,
                   2: 500,
                   3: 1000,
                   4: 10000}

def rate_mortality(subtable):
  mortality = subtable["Deaths"]
  if (mortality <= mortality_scale[0]):
    return 0
  elif (mortality <= mortality_scale[1]):
    return 1
  elif (mortality <= mortality_scale[2]):
    return 2
  elif (mortality <= mortality_scale[3]):
    return 3
  elif mortality <= mortality_scale[4]:
    return 4
  else:
    return 5

def create_mortality_table(hurricane_table):
  mortality_table = {0: [], 1: [], 2: [], 3: [], 4: [], 5: []}
  for hurricane in hurricane_table.values():
    rating = rate_mortality(hurricane)
    mortality_table[rating].append(hurricane)
  return mortality_table

mortality_table = create_mortality_table(hurricane_table)
mortality_table[4]
```

```python
def costliest(hurricane_table):
  current_max_value = 0
  current_max_key = ''
  for subtable in hurricane_table.values():
    current_key = subtable["Name"]
    current_value = damage_to_float(subtable["Damage"])
    if current_value == "Damages not recorded":
      continue
    if current_value > current_max_value:
      current_max_value = current_value
      current_max_key = current_key
    elif current_value == current_max_value:
      current_max_key += ", " + current_key
  return current_max_key, current_max_value

costliest(hurricane_table)
```

```python
damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}

def rate_damage(subtable):
  damage = damage_to_float(subtable["Damage"])
  if damage == 'Damages not recorded':
    return damage
  if (damage <= damage_scale[0]):
    return 0
  elif (damage <= damage_scale[1]):
    return 1
  elif (damage <= damage_scale[2]):
    return 2
  elif (damage <= damage_scale[3]):
    return 3
  elif damage <= damage_scale[4]:
    return 4
  else:
    return 5

def create_damage_table(hurricane_table):
  damage_table = {0: [], 1: [], 2: [], 3: [], 4: [], 5: []}
  for hurricane in hurricane_table.values():
    rating = rate_damage(hurricane)
    if rating == 'Damages not recorded':
      continue
    damage_table[rating].append(hurricane)
  return damage_table

damage_table = create_damage_table(hurricane_table)
damage_table[5]
```
